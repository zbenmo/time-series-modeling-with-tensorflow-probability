```python
min_seen = 14
max_seen = 222

with pm.Model() as model:
    mu = pm.Normal(name="mu", mu=(min_seen + max_seen) / 2, sigma=(max_seen - min_seen) / 4)
    sigma = pm.HalfNormal(name="sigma", sigma=10)
    y = pm.Normal(name="y", mu=mu, sigma=sigma, observed=df.IsCanceled)
```

```python
with model:
    
    trace = pm.sample(1000, tune=2000, random_seed=RANDOM_SEED, return_inferencedata=True)
#     idata = az.from_pymc3(trace)
```
